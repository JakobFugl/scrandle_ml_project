```python
import os
import hashlib
```

```python
def hash_file(path):
    with open(path, "rb") as f:
        return hashlib.md5(f.read()).hexdigest()
```

```python
base_folder = "./scrandle_data"

date_sequences = {}

for date_folder in os.listdir(base_folder):
    full_path = os.path.join(base_folder, date_folder)

    if not os.path.isdir(full_path):
        continue

    # IMPORTANT: enforce correct order
    images = sorted(
        [f for f in os.listdir(full_path) if f.endswith(".webp")],
        key=lambda x: (
            int(x.split("_")[0]),   # match_id
            0 if "left" in x else 1 # left before right
        )
    )

    # Hash images IN ORDER
    sequence = []
    for img in images:
        img_path = os.path.join(full_path, img)
        sequence.append(hash_file(img_path))

    date_sequences[date_folder] = tuple(sequence)
```

```python
from collections import defaultdict

sequence_groups = defaultdict(list)

for date, seq in date_sequences.items():
    sequence_groups[seq].append(date)

```

```python
found = False

for seq, dates in sequence_groups.items():
    if len(dates) > 1:
        found = True
        print("⚠️ IDENTICAL IMAGE SEQUENCE FOUND:")
        print(dates)
        print("-" * 50)

if not found:
    print("✅ No identical image sequences found.")
```
