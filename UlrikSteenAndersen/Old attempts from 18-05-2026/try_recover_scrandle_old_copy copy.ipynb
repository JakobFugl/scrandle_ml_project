```python
url = "https://scrandle.com/history/2026-05-03"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

data = response.json()   # ✅ THIS WORKS HERE
```

```python
results = []

for pair in data["data"]:
    left, right = pair

    entry = {
        "left": {
            "title": left["title"],
            "subtitle": left["subtitle"],
            "price": left["price"],
            "rating": left["rating"],
            "image": left["images_new"][0]
        },
        "right": {
            "title": right["title"],
            "subtitle": right["subtitle"],
            "price": right["price"],
            "rating": right["rating"],
            "image": right["images_new"][0]
        }
    }

    results.append(entry)

for r in results:
    print(r)
```

```python
url = "https://scrandle.com/history/2026-05-03"
headers = {"User-Agent": "Mozilla/5.0"}

data = requests.get(url, headers=headers).json()

results = []

for left, right in data["data"]:
    entry = {
        "left": {
            "title": left["title"],
            "subtitle": left["subtitle"],
            "price": left["price"],
            "rating": left["rating"],
            "image": left["images_new"][0]
        },
        "right": {
            "title": right["title"],
            "subtitle": right["subtitle"],
            "price": right["price"],
            "rating": right["rating"],
            "image": right["images_new"][0]
        }
    }
    results.append(entry)

# Print nicely
for r in results:
    print("\n---")
    print(r)
```

```python
os.makedirs("images", exist_ok=True)

for i, r in enumerate(results):
    for side in ["left", "right"]:
        img_url = r[side]["image"]
        img_data = requests.get(img_url).content
        
        filename = f"images/{i}_{side}.webp"
        with open(filename, "wb") as f:
            f.write(img_data)
```
