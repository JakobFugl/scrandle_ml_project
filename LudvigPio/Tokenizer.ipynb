```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
import pandas as pd
import os
import numpy as np
```

```python
def tokenizer(text_list):
    tokenizer = Tokenizer(num_words=10*len(text_list[0]), oov_token="<OOV>")
    tokenizer.fit_on_texts(text_list)
    sequences = tokenizer.texts_to_sequences(text_list)
    return sequences, tokenizer.word_index
```

```python
def data_processor(path):
    data = {}
    date_list = sorted(os.listdir(path))
    title_list = []
    subtitle_list = []
    club_list = []
    country_list = []
    for date in date_list:
        meta_path = os.path.join(path, date, "meta.csv")
        data[date] = {}
        data[date] = pd.read_csv(meta_path,sep=",").fillna("")
        for index, row in data[date].iterrows():
            title_list.append(row["title"])
            subtitle_list.append(row["subtitle"])
            club_list.append(row["club"])
            country_list.append(row["country"])
    return data, title_list, subtitle_list, club_list, country_list
```

```python
def main(path):
    data,title_list, subtitle_list, club_list, country_list = data_processor(path)
    text_list = [title_list, subtitle_list, club_list, country_list]
    sequences, tokenizer_word_index = tokenizer(text_list)
    return data, title_list, subtitle_list, text_list, sequences, tokenizer_word_index
```

```python
path = r"..\data\grand_scraper_folder\scrandle_data"

data, title_list, subtitle_list, text_list, sequences, tokenizer_word_index = main(path)

print(text_list)
print(sequences)
print(tokenizer_word_index)
```
